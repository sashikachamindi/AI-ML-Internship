```python
import pandas as pd
```

```python
data = pd.DataFrame({
    "name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ian","Jack","Kate"],
    "result": ["Pass","Fail","Pass","Pass","Absent","Fail","Fail","Pass","Pass","Absent","Pass"]
})
```

```python
print(data['result'])
```

```python
result_category = data ['result']
```

```python
from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder ()
result = obj.fit_transform (result_category)
```

```python
result
```

```python
from sklearn.preprocessing import LabelBinarizer
obj = LabelBinarizer ()
result = obj.fit_transform (result_category)
```

```python
result
```
