```python
# Numerical computing
import numpy as np  

# Plotting graphs
import matplotlib.pyplot as plt  

# Math functions (optional if using numpy)
import math 
```

```python
# -----------------------------
# Example 1: Simple Linear Regression
# -----------------------------

# Training data
X = np.array([5, 6, 7, 8, 9, 10, 11])
y = np.array([3, 6, 4, 8, 9, 12, 16])
```

```python
# Initialize slope (m) and intercept (b)
m = 0
b = 0

# Learning rate and iterations
alpha = 0.01
iterations = 1000
n = len(X)
```

```python
# Gradient Descent
for i in range(iterations):
    # Compute predicted values
    y_pred = m*X + b
    
    # Compute cost (Mean Squared Error)
    J = (1/n) * sum((y_pred - y)**2)
    
    # Compute gradients
    dm = (2/n) * sum((y_pred - y) * X)
    db = (2/n) * sum(y_pred - y)
    
    # Update parameters
    m = m - alpha * dm
    b = b - alpha * db
    
    print(f"Iteration {i}: cost={J}")

print(f"Final slope (m) = {m}, intercept (b) = {b}")
```

```python
# -----------------------------
# Example 2: Another Linear Regression
# -----------------------------

# New training data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 8, 11, 14, 17])
```

```python
# Initialize parameters
a = 0  # slope
b = 0  # intercept
learning_rate = 0.01
n = len(x)
```

```python
# Gradient Descent
for i in range(100):
    y_pred = a*x + b
    
    # Compute cost (MSE)
    cost = (1/n) * sum((y - y_pred)**2)
    
    # Plot cost for visualization
    plt.scatter(i, cost, color='blue')
    
    # Compute gradients
    da = (2/n) * sum(x * (y - y_pred))
    db = (2/n) * sum(y - y_pred)
    
    # Update parameters
    a = a - learning_rate * da
    b = b - learning_rate * db
    
    # Stop if cost is very small
    if math.isclose(cost, 0, abs_tol=1e-6):
        break

plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Reduction Over Iterations")
plt.show()

print(f"Final slope (a) = {a}, intercept (b) = {b}")
```
